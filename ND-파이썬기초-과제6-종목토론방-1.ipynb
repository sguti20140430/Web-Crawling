### 페이지별 데이터 수집 함수 만들기

```python
def get_day_list(item_code, page_no):
    """
    일자별 게시글을 페이지별로 수집
    """ 
    url = f'https://finance.naver.com/item/board.nhn?code={item_code}&page={page_no}'
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36'}
    response = requests.get(url, headers=headers)
    html = bs(response.text, 'lxml')
    temp = html.select('table')
    df_day = pd.read_html(str(temp))[1]
    df_day = df_day.dropna(axis=0, how='all').dropna(axis=1)
    return df_day
```

```python
page_no = 10
item_code = "005930"
get_day_list(item_code, page_no)
```
